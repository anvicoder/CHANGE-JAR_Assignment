```python

import matplotlib.pyplot as plt



# Create a boxplot for the "Unit price" column
plt.boxplot(dataset["Unit price"])

# Add a title and labels
plt.title("Boxplot of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Value")

# Display the boxplot
plt.show()
```

```python

import matplotlib.pyplot as plt



# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

# Plot Branch
dataset["Branch"].value_counts().plot(kind="bar", ax=axes[0, 0])
axes[0, 0].set_title("Branch")

# Plot City
dataset["City"].value_counts().plot(kind="bar", ax=axes[0, 1])
axes[0, 1].set_title("City")

# Plot Customer Type
dataset["Customer type"].value_counts().plot(kind="bar", ax=axes[1, 0])
axes[1, 0].set_title("Customer Type")

# Plot Gender
dataset["Gender"].value_counts().plot(kind="bar", ax=axes[1, 1])
axes[1, 1].set_title("Gender")

# Plot Product Line
dataset["Product line"].value_counts().plot(kind="bar", ax=axes[2, 0])
axes[2, 0].set_title("Product Line")

# Plot Payment
dataset["Payment"].value_counts().plot(kind="bar", ax=axes[2, 1])
axes[2, 1].set_title("Payment")

# Improve layout and display
plt.tight_layout()
plt.show()
```

```python
# Calculate the sales by multiplying the price by the quantity
dataset['Sales'] = dataset['Unit price'] * dataset['Quantity']
print (dataset['Sales'])
total_sales = dataset['Sales'].sum()
print(f"Total sales: {total_sales}")
```

```python
# Calculating average price
total_sales = dataset['Sales'].sum()
num_sales = len(dataset)
average_price = total_sales / num_sales
print(f"Average price: {average_price}")
```

```python
# Calculating the revenue by multiplying the sales by the avg, sales price
dataset['Revenue'] = dataset['Sales'] * average_price
print (dataset['Revenue'])
total_sales = dataset['Revenue'].sum()
print(f"Total Revenue: {total_sales}")
```

```python
dataset['Revenue'] = dataset['Revenue'].round(2)

# Alternatively, using apply() function with a lambda function
dataset['Revenue'] = dataset['Revenue'].apply(lambda x: round(x, 2))

# Display the DataFrame with reduced decimal places
print(dataset['Revenue'])
```

```python
grouped_data = dataset.groupby(['City', 'Branch'])
```

```python
sales_by_city_branch = grouped_data['Sales'].sum()
revenue_by_city_branch = grouped_data['Revenue'].sum()
```

```python
avg_sales_by_city_branch = grouped_data['Sales'].mean()
avg_revenue_by_city_branch = grouped_data['Revenue'].mean()
```

```python
import matplotlib.pyplot as plt

# Create bar charts
plt.figure(figsize=(10, 6))
sales_by_city_branch.plot(kind='bar')
plt.title('Sales by City and Branch')
plt.xlabel('City and Branch')
plt.ylabel('Sales')
plt.show()

plt.figure(figsize=(10, 6))
revenue_by_city_branch.plot(kind='bar')
plt.title('Revenue by City and Branch')
plt.xlabel('City and Branch')
plt.ylabel('Revenue')
plt.show()

```

# C. Analyze the performance of sales and revenue, Month over Month across the Product line, Gender, and Payment Method, and identify the focus areas to get better sales for April 2019




```python


# Extract the month from the date column
dataset['Month'] = pd.to_datetime(dataset['Date']).dt.month

# Print the DataFrame with the new month column
print(dataset['Month'])
```

```python
grouped_data = dataset.groupby(['Product line', 'Gender', 'Payment', 'Month']).agg({'Sales': 'sum', 'Revenue': 'sum'}).reset_index()
print(grouped_data)
```

```python
df_filtered = dataset[["Product line", "Gender", "Payment", "Month", "Sales", "Revenue"]]

# Group data
grouped_data = df_filtered.groupby(["Month", "Product line", "Gender", "Payment"])[["Sales", "Revenue"]].sum()

# Analyze performance
# 1. Analyze sales and revenue trends over time for each grouping
# 2. Compare sales and revenue across different groupings
# 3. Identify underperforming product lines, genders, or payment methods

# Identify focus areas
# 1. Investigate potential reasons for underperformance
# 2. Suggest strategies to improve sales for April 2019

# Visualization
# Create visualizations to present the analysis results

# Plot average sales per month for each product line
plt.figure(figsize=(10, 8))
grouped_data.groupby(["Month", "Product line"])["Sales"].mean().unstack().plot(kind="bar")
plt.title("Average Sales per Month by Product Line")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.show()
# Plot revenue per month for each product line
plt.figure(figsize=(10, 8))
grouped_data.groupby(["Month", "Product line"])["Revenue"].mean().unstack().plot(kind="bar")
plt.title("Average Revenue per Month by Product Line")
plt.xlabel("Month")
plt.ylabel("Average Revenue")
plt.show()


```

```python
# Plot total sales per month for each gender
plt.figure(figsize=(10, 7))
grouped_data.groupby(["Month", "Gender"])["Sales"].sum().unstack().plot(kind="bar")
plt.title("Total Sales per Month by Gender")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
# Plot total revenue per month for each gender
plt.figure(figsize=(10, 7))
grouped_data.groupby(["Month", "Gender"])["Revenue"].sum().unstack().plot(kind="bar")
plt.title("Total Revenue per Month by Gender")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Plot total sales per month for each gender
plt.figure(figsize=(10, 7))
grouped_data.groupby(["Month", "Payment"])["Sales"].sum().unstack().plot(kind="bar")
plt.title(" Monthly  Payment")
plt.xlabel("Month")
plt.ylabel("Total Payment")
plt.show()
# Plot total revenue per month for each gender
plt.figure(figsize=(10, 7))
grouped_data.groupby(["Month", "Payment"])["Revenue"].sum().unstack().plot(kind="bar")
plt.title(" Monthly  Payment")
plt.xlabel("Month")
plt.ylabel("Total Payment")
plt.show()
```

# We need to increase the sales of food and beverages in order to get better sales for April 2019 i.e. 4th month.

```python
# Group data by month and product line
monthly_sales = dataset.groupby(["Month", "Product line"])["Sales"].sum().reset_index()

# Calculate the average sales for each product line
average_sales = monthly_sales.groupby("Product line")["Sales"].mean().sort_values(ascending=False)

# Identify the product line with the highest average sales
highest_average_sales_product_line = average_sales.index[0]

print(f"Product Line with the Highest Average Sales per Month: {highest_average_sales_product_line}")
```
